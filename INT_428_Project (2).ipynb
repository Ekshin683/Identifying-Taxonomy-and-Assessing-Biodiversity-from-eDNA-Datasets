```python
import pandas as pd


occurrence_path = "/content/occurrence.csv"
dna_path = "/content/dna_extension.csv"

occ = pd.read_csv(occurrence_path)
dna = pd.read_csv(dna_path)


print(" occurrence.csv shape:", occ.shape)
print(" dna_extension.csv shape:", dna.shape)

print("\n occurrence.csv columns:")
print(list(occ.columns))

print("\n dna_extension.csv columns:")
print(list(dna.columns))


required_keys = ["eventID", "occurrenceID"]

for key in required_keys:
    if key not in occ.columns:
        print(f" '{key}' not found in occurrence.csv")
    if key not in dna.columns:
        print(f" '{key}' not found in dna_extension.csv")


df = occ.merge(dna, on=["eventID", "occurrenceID"], how="inner")


print("\n MERGED DATASET SHAPE:", df.shape)

print("\n Merged dataset preview:")
df.head()
```

```python

import numpy as np
import re
import matplotlib.pyplot as plt


df_clean = df.copy()


if "DNA_sequence" not in df_clean.columns:
    raise ValueError("'DNA_sequence' column not found in your dataset!")

print(" DNA_sequence column found.")


before_rows = df_clean.shape[0]
df_clean = df_clean.dropna(subset=["DNA_sequence"])
after_rows = df_clean.shape[0]

print(f" Removed {before_rows - after_rows} rows with missing DNA.")


df_clean["DNA_sequence"] = df_clean["DNA_sequence"].astype(str).str.upper()

#  Remove invalid characters (keep only A, C, G, T)
def clean_dna(seq):
    return re.sub(r"[^ACGT]", "", seq)

df_clean["DNA_sequence"] = df_clean["DNA_sequence"].apply(clean_dna)


df_clean["seq_length"] = df_clean["DNA_sequence"].str.len()

min_length = 80
df_clean = df_clean[df_clean["seq_length"] >= min_length]

print(" After filtering short sequences:", df_clean.shape)


df_clean = df_clean.reset_index(drop=True)


plt.figure()
plt.hist(df_clean["seq_length"], bins=50)
plt.title("DNA Sequence Length Distribution (After Cleaning)")
plt.xlabel("Sequence Length")
plt.ylabel("Frequency")
plt.show()


print("\n FINAL CLEAN DATASET SUMMARY:")
print(df_clean[["DNA_sequence", "seq_length"]].head())
print("\n Clean dataset shape:", df_clean.shape)
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer


if "df_clean" not in globals():
    raise ValueError(" df_clean not found! Run Phase 2 first.")

print(" df_clean found with shape:", df_clean.shape)

#  Choose k-mer size
k = 5

print(f" Using k-mer size = {k}")


def seq_to_kmers(seq, k):
    return [seq[i:i+k] for i in range(len(seq) - k + 1)]


dna_docs = df_clean["DNA_sequence"].apply(lambda x: " ".join(seq_to_kmers(x, k)))

vectorizer = CountVectorizer(token_pattern=r"(?u)\b\w+\b")
X_kmer = vectorizer.fit_transform(dna_docs)


print(" k-mer Feature Matrix Shape:", X_kmer.shape)
print(" Total unique k-mers:", len(vectorizer.get_feature_names_out()))


X_kmer_preview = X_kmer[:5].toarray()
kmer_features = vectorizer.get_feature_names_out()

kmer_df_preview = pd.DataFrame(X_kmer_preview, columns=kmer_features)
print("\n k-mer Feature Preview (first 5 sequences):")
kmer_df_preview.iloc[:, :10].head()
```

```python
from sklearn.decomposition import TruncatedSVD


if "X_kmer" not in globals():
    raise ValueError(" X_kmer not found! Run Phase 3 first.")
if "df_clean" not in globals():
    raise ValueError(" df_clean not found!")

print("X_kmer shape:", X_kmer.shape)


n_components_svd = 20

svd = TruncatedSVD(n_components=n_components_svd, random_state=42)
X_svd_fast = svd.fit_transform(X_kmer)

print(" Reduced shape:", X_svd_fast.shape)
print(" Total explained variance:", svd.explained_variance_ratio_.sum())


for i in range(n_components_svd):
    df_clean[f"SVD_{i+1}"] = X_svd_fast[:, i]

df_clean[[f"SVD_{i+1}" for i in range(5)]].head()
```

```python
import numpy as np
import pandas as pd


if "df_clean" not in globals():
    raise ValueError(" df_clean not found!")

svd_cols = [col for col in df_clean.columns if col.startswith("SVD_")]
if len(svd_cols) == 0:
    raise ValueError(" SVD features not found! Run Fast Phase 4 first.")

print(" Found SVD features:", len(svd_cols))


X_cluster = df_clean[svd_cols].values
print("Clustering feature matrix shape:", X_cluster.shape)


try:
    import hdbscan
except ImportError:
    import hdbscan


clusterer = hdbscan.HDBSCAN(
    min_cluster_size=25,
    min_samples=10,
    metric='euclidean',
    cluster_selection_method='eom'
)

labels = clusterer.fit_predict(X_cluster)
df_clean["cluster"] = labels


n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = np.sum(labels == -1)

print("\n CLUSTER SUMMARY")
print(" Total data points:", len(labels))
print(" Total clusters found:", n_clusters)
print(" Noise / outliers:", n_noise)

# Show cluster size distribution (top 10)
cluster_sizes = pd.Series(labels).value_counts().head(10)
print("\n Top 10 cluster sizes:")
print(cluster_sizes)

# Preview clustered data
df_clean[["DNA_sequence", "seq_length", "cluster"]].head()
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import (
    adjusted_rand_score,
    normalized_mutual_info_score,
    silhouette_score
)
required_cols = ["cluster", "scientificName", "genus"]
for col in required_cols:
    if col not in df_clean.columns:
        raise ValueError(f" Required column '{col}' not found in df_clean!")

svd_cols = [col for col in df_clean.columns if col.startswith("SVD_")]
if len(svd_cols) == 0:
    raise ValueError(" SVD feature columns not found!")
df_eval = df_clean[df_clean["cluster"] != -1].copy()
print(" Evaluation dataset shape (without noise):", df_eval.shape)

species_labels, _ = pd.factorize(df_eval["scientificName"])
genus_labels, _ = pd.factorize(df_eval["genus"])
cluster_labels = df_eval["cluster"].values


ari_species = adjusted_rand_score(species_labels, cluster_labels)
nmi_species = normalized_mutual_info_score(species_labels, cluster_labels)

ari_genus = adjusted_rand_score(genus_labels, cluster_labels)
nmi_genus = normalized_mutual_info_score(genus_labels, cluster_labels)

print("\n EXTERNAL CLUSTER VALIDATION METRICS")
print(" ARI (Scientific Name):", round(ari_species, 4))
print(" NMI (Scientific Name):", round(nmi_species, 4))
print(" ARI (Genus):", round(ari_genus, 4))
print(" NMI (Genus):", round(nmi_genus, 4))


purity_list = []

for cluster_id, group in df_eval.groupby("cluster"):
    total = len(group)
    majority_count = group["scientificName"].value_counts().iloc[0]
    purity = majority_count / total
    purity_list.append(purity)

avg_purity = np.mean(purity_list)

print("\n Average Cluster Purity (Species-level):", round(avg_purity, 4))

cluster_species_map = []

for cluster_id, group in df_eval.groupby("cluster"):
    dominant_species = group["scientificName"].value_counts().idxmax()
    dominant_count = group["scientificName"].value_counts().iloc[0]
    total = len(group)

    cluster_species_map.append({
        "cluster": cluster_id,
        "dominant_species": dominant_species,
        "dominant_count": dominant_count,
        "cluster_size": total,
        "purity": dominant_count / total
    })

cluster_species_df = pd.DataFrame(cluster_species_map)

print("\n Cluster → Dominant Species Mapping (Top 10 by size):")
cluster_species_df.sort_values(by="cluster_size", ascending=False).head(10)


X_sil = df_eval[svd_cols].values
labels_sil = df_eval["cluster"].values

#  Sample for speed (important for large datasets)
max_samples = 5000
if X_sil.shape[0] > max_samples:
    idx = np.random.choice(X_sil.shape[0], max_samples, replace=False)
    X_sil = X_sil[idx]
    labels_sil = labels_sil[idx]
    print(f" Silhouette score computed on random sample of {max_samples} points")

sil_score = silhouette_score(X_sil, labels_sil, metric="euclidean")

print("\n INTERNAL CLUSTER QUALITY METRIC")
print(" Silhouette Score:", round(sil_score, 4))
cluster_species_df.to_csv("cluster_species_mapping.csv", index=False)

metrics_df = pd.DataFrame({
    "Metric": [
        "ARI (Species)",
        "NMI (Species)",
        "ARI (Genus)",
        "NMI (Genus)",
        "Average Cluster Purity",
        "Silhouette Score"
    ],
    "Value": [
        ari_species,
        nmi_species,
        ari_genus,
        nmi_genus,
        avg_purity,
        sil_score
    ]
})

metrics_df.to_csv("clustering_evaluation_metrics.csv", index=False)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform

required_cols = ["eventID", "cluster"]
for col in required_cols:
    if col not in df_clean.columns:
        raise ValueError(f" Required column '{col}' not found!")

print(" Required columns found for biodiversity analysis.")

df_bio = df_clean[df_clean["cluster"] != -1].copy()
print(" Biodiversity dataset shape (no noise):", df_bio.shape)

abundance_table = pd.crosstab(df_bio["eventID"], df_bio["cluster"])

print("\n Abundance Table Shape (events × clusters):")
print(abundance_table.shape)

abundance_table.to_csv("event_cluster_abundance_table.csv")
print("Saved: event_cluster_abundance_table.csv")

# Alpha Diversity

def shannon_index(counts):
    counts = counts[counts > 0]
    proportions = counts / counts.sum()
    return -np.sum(proportions * np.log(proportions))

def simpson_index(counts):
    counts = counts[counts > 0]
    proportions = counts / counts.sum()
    return 1 - np.sum(proportions**2)

alpha_results = []

for event_id, row in abundance_table.iterrows():
    counts = row.values

    richness = np.sum(counts > 0)
    shannon = shannon_index(counts)
    simpson = simpson_index(counts)

    alpha_results.append({
        "eventID": event_id,
        "richness": richness,
        "shannon_index": shannon,
        "simpson_index": simpson
    })

alpha_df = pd.DataFrame(alpha_results)

# Save alpha diversity
alpha_df.to_csv("alpha_diversity_results.csv", index=False)
print(" Saved: alpha_diversity_results.csv")

print("\n ALPHA DIVERSITY PREVIEW:")
alpha_df.head()

#Beta diversity

def bray_curtis_distance(u, v):
    return np.sum(np.abs(u - v)) / np.sum(u + v)
beta_distance_matrix = squareform(
    pdist(abundance_table.values, metric=bray_curtis_distance)
)

beta_df = pd.DataFrame(
    beta_distance_matrix,
    index=abundance_table.index,
    columns=abundance_table.index
)
beta_df.to_csv("beta_diversity_bray_curtis.csv")
print(" Saved: beta_diversity_bray_curtis.csv")

from sklearn.manifold import MDS

mds = MDS(n_components=2, dissimilarity="precomputed", random_state=42)
pcoa_coords = mds.fit_transform(beta_distance_matrix)

pcoa_df = pd.DataFrame(
    pcoa_coords,
    columns=["PCoA1", "PCoA2"],
    index=abundance_table.index
)
pcoa_df.to_csv("pcoa_coordinates.csv")
print(" Saved: pcoa_coordinates.csv")

plt.figure()
plt.hist(alpha_df["shannon_index"], bins=30)
plt.title("Shannon Diversity Distribution Across Events")
plt.xlabel("Shannon Index")
plt.ylabel("Number of Events")
plt.show()

plt.figure()
plt.scatter(pcoa_df["PCoA1"], pcoa_df["PCoA2"])
plt.title("PCoA Plot Based on Bray–Curtis Beta Diversity")
plt.xlabel("PCoA1")
plt.ylabel("PCoA2")
plt.show()

plt.figure(figsize=(10, 7))


cluster_plot_data = df_clean[df_clean["cluster"] != -1]

scatter = plt.scatter(
    cluster_plot_data["SVD_1"],
    cluster_plot_data["SVD_2"],
    c=cluster_plot_data["cluster"],
    cmap="tab20",
    s=6,
    alpha=0.7
)

plt.colorbar(scatter, label="Cluster ID")
plt.title("Cluster Visualization (SVD Space)")
plt.xlabel("SVD 1")
plt.ylabel("SVD 2")
plt.show()
```

```python
import numpy as np
import pandas as pd
import re
from sklearn.neighbors import NearestCentroid
def get_user_sequence():
    seq = input("Enter DNA sequence (A,C,G,T): ").upper()
    seq = re.sub(r"[^ACGT]", "", seq)
    return seq

def seq_to_kmers(seq, k=5):
    if len(seq) < k:
        raise ValueError("\u274c DNA sequence too short for k-mer extraction!")
    kmers = [seq[i:i+k] for i in range(len(seq)-k+1)]
    return " ".join(kmers)


def predict_cluster(user_seq):


    kmer_doc = seq_to_kmers(user_seq, k=5)


    X_user_kmer = vectorizer.transform([kmer_doc])


    X_user_svd = svd.transform(X_user_kmer)

    global centroid_clf

    if "centroid_clf" not in globals():
        valid_df = df_clean[df_clean["cluster"] != -1]
        X_train = valid_df[svd_cols].values
        y_train = valid_df["cluster"].values

        centroid_clf = NearestCentroid()
        centroid_clf.fit(X_train, y_train)
        print("✔ Centroid classifier trained for prediction.")

    predicted_cluster = centroid_clf.predict(X_user_svd)[0]

    if predicted_cluster in df_clean["cluster"].unique():
        sub = df_clean[df_clean["cluster"] == predicted_cluster]


        top_species = sub["scientificName"].value_counts().head(3)
        species_text = "; ".join(
            [f"{name} ({count})" for name, count in top_species.items()]
        )


        if "genus" in sub.columns:
            genus_counts = sub["genus"].value_counts()
            top_genus = genus_counts.idxmax() if not genus_counts.empty else "Unknown"
        else:
            top_genus = "Unknown"

        if "family" in sub.columns:
            family_counts = sub["family"].value_counts()
            top_family = family_counts.idxmax() if not family_counts.empty else "Unknown"
        else:
            top_family = "Unknown"

    else:
        species_text = "Unknown"
        top_genus = "Unknown"
        top_family = "Unknown"

    return predicted_cluster, species_text, top_genus, top_family



user_seq = get_user_sequence()

cluster_id, species_text, genus_text, family_text = predict_cluster(user_seq)

print("\n==============================")
print("FINAL PREDICTION RESULT")
print("==============================")
print("Predicted Cluster ID:", cluster_id)
print("Likely Species:", species_text)
print("Likely Genus:", genus_text)
print("Likely Family:", family_text)
print("==============================")
```

```python
!pip install -q gradio

import gradio as gr
import re
import numpy as np
from sklearn.neighbors import NearestCentroid

required_objects = ["df_clean", "vectorizer", "svd", "svd_cols"]
for obj in required_objects:
    if obj not in globals():
        raise ValueError(f"❌ Required object '{obj}' not found. Run previous phases first.")


valid_df = df_clean[df_clean["cluster"] != -1]

X_train = valid_df[svd_cols].values
y_train = valid_df["cluster"].values

centroid_clf = NearestCentroid()
centroid_clf.fit(X_train, y_train)

print("✅ Centroid classifier ready")

def clean_dna(seq):
    return re.sub(r"[^ACGT]", "", seq.upper())

def seq_to_kmers(seq, k=5):
    return " ".join(seq[i:i+k] for i in range(len(seq)-k+1))

def predict_dna(dna_sequence):

    try:
        dna_sequence = clean_dna(dna_sequence)

        if len(dna_sequence) < 80:
            return "❌ Too short", "-", "-", "-"

        # Convert to k-mers
        kmer_doc = seq_to_kmers(dna_sequence)
        X_kmer = vectorizer.transform([kmer_doc])
        X_svd = svd.transform(X_kmer)

        # Predict cluster
        cluster = int(centroid_clf.predict(X_svd)[0])

        # Check if cluster exists in dataset
        sub = df_clean[df_clean["cluster"] == cluster]

        if sub.empty:
            return (
                str(cluster),
                "Unknown (no dominant species)",
                "Unknown",
                "Unknown"
            )

        # Safe taxonomy extraction
        species = "; ".join(
            [f"{k} ({v})" for k, v in sub["scientificName"].value_counts().head(3).items()]
        )

        genus = (
            sub["genus"].value_counts().idxmax()
            if "genus" in sub.columns and not sub["genus"].dropna().empty
            else "Unknown"
        )

        family = (
            sub["family"].value_counts().idxmax()
            if "family" in sub.columns and not sub["family"].dropna().empty
            else "Unknown"
        )

        return str(cluster), species, genus, family

    except Exception as e:
        return f"❌ Error: {str(e)}", "-", "-", "-"



demo = gr.Interface(
    fn=predict_dna,
    inputs=gr.Textbox(lines=4, placeholder="Enter DNA sequence (A,C,G,T only)"),
    outputs=[
        gr.Textbox(label="Predicted Cluster ID"),
        gr.Textbox(label="Likely Species"),
        gr.Textbox(label="Likely Genus"),
        gr.Textbox(label="Likely Family")
    ],
    title="🧬 eDNA Taxonomy & Biodiversity Predictor",
    description="Enter an eDNA sequence to predict its taxonomic cluster using an unsupervised ML model."
)

demo.launch(share=True, debug=True)
```
